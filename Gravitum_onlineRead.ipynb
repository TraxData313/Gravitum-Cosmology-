```python
# Appendix:
#
# - Recently a deep connection between Quantum Entanglement Complexity 
#   and the Structure of Space has become a very hot subject of study.
# - There are a lot of technical papers online regarding this subject,
#   but there are little available to the general public, 
#   on this new and exciting topic.
# - I decided to put toghether this short Jupyter note,
#   where we'll just follow the consequences of a powerful assumption:
#   "Space IS nothing but Quantum Complexity"

# So out plan of action will be:
#
# 1. Count how many bits of information there are in our Universe:
#  - Take the observed Universe and crush it into a Black Hole.
#    In order to answer how much bits there are in out Universe,
#    we will count the pixels on the Horizon of the resulting Blach Hole.

# 2. Calculate the Quantum Complexity of our Universe' bits:
#  - Using the Second Law of Quantum Complexity, we are going to
#    calculate how the complexity of those bits evolves with time.
#    SPOILER: It grows fast!

# 3. After that we are going to take the hint from the Holographic
#    principle (Space = Quantum Complexity), and are going to 
#    assume that Growth of Quantum Complexity = Growth of Space

# 4. Once here, we will have our model of the Universe expansion and put it to the test:
#  - We are going to calculate the predicted Hubble expansion rate for our
#    theory, and compare it with the real (observed) Hubble exp rate

# 5. Now, let's plot our model agains the data from the standard expansion models:
#  - Currently our Universe expansion is modeled by the Big Bang Theory, which
#    can be traced back to almost time = 0 seconds.
#  - Before that, Inflation theory predicts that the Universe expanded rapidly in a fraction
#    of a second, just moments after time = 0 (at about t = 1e-35 seconds (0.0...35zeroshere..01 seconds))
#  - There is no "standard" theory of what happened before Inflation.

#  - Will our model be able to extend back to that "before Inflation" time?
#  - Will our model show a smoother picture of what could have happened?
#  - Can we go even further?

# Lets get started and see!

# What we you need to be able to follow this presentation?
# - Only some hight school maths and maybe basic Python knowedge
# - We are going to let Python do all the heavy leafting for us
```

```python
# CONSTANTS we will use, taken from Wikipedia:

# - Cosmological observables:
# -- Universe Mass:
UM = 3.33E+53 #[kg]
# -- Universe Volume:
UV = 4.00E+80 #[m^3]
# -- Universe Radius:
UR = 4.40E+26 #[m]
# -- Universe Life:
UL = 4.35E+17 #[sec]
# -- Hubble exp rate in SI units:
H = 2.35E-18 #[(m/sec)/m]

# -- Speed of light:
c = 3.00E+08 #[m/s]
# -- Grav. constant:
G = 6.67E-11
# -- Pi:
pi = 3.14

# - Plank units in SI:
# -- Plank lenght:
Lp = 1.616E-35   #[m]
# -- Plank area:
Ap = 2.61219E-70 #[m^2]
# -- Plank Volume:
Vp = 4.2217E-105 #[m^3]
# -- Plank time:
Tp = 5.39116E-44 #[sec]
```

```python
# From Black Hole (BH) thermodynamics we know that one bit 
# takes exactly one Plank Area on the BH horizon

# So lets use that fact to determine how much bits exist
# in out universe. Lets crush the universe in a BH and count.

# What is the radius (Schwarzchild radius) of a black hole 
# as a function of its mass? Here is the equation:

# Radius = 2*G*Mass/c^2 [m]
#   Where:
#   - G: Grav. Constant
#   - c: speed of light

# Lets plug in this equation the Universe Mass and calculate the 
# Universe Radius if it was a Black Hole:

Radius = 2*G*UM/(c**2)

# Lets print it now:
print("Universe Blach Hole Radius:", Radius, "[m^2]")
```

```python
# In order to count the bits, we need the area of the BH horizon,
# we will use the geometrical formula for the Area of a Sphere
# as a function of its radius, here it is:

# Area = 4*pi*Radius^2

# And let's calculate the Blach Hole Horizon Area of out Universe
# if it was a Black Hole:
Area = 4*pi*(Radius**2)

print("Universe Black Hole Area:", Area, "[m^2]")
```

```python
# - How much bits of information make up a Blach Hole?
#   There is 1 bit for every Plank Area on the BH Horizon Area
#   So number of bits = Total Area / Plank Area

UB = Area / Ap

# FYI: If you are wondering exactly how scientists found out how much
#      bits of pure information there are in a Black Hole
#      you can read more on the subject of "Black hole thermodynamics"

# - But now, lets just print them out:
print("Our Universe has:", UB, "bits")
```

```python
# One of the most fundamental physics principles is that information
# never disappears and never gets created, so once we 
# calculate how much bits our Universe would have if it was a BH,
# we know that it would have the same number of bits in its every state, even now

#  UB = const

# Long story short, we now know how much bits of information make up
# our Universe, given the Mass we observe,
# and that in just a few lines of code!
```

```python
# GROWTH OF THE QUANTUM COMPLEXITY

# What is Complexity in quantum information theory?

# - Quantum bits entangle with one another, one can say that
#   they create connections. The Complexity represents the
#   number of those connections. The more connected the system
#   is - the more entangled, or Complex, it gets.

# The Second Law of Quantum Complexity tells us how fast the
# Complexity of a system, with a given amount of bits grows with time.
# Complexity as a function of time C(t) equation is:

# C(t) = Bits*Time

# So if the system is made from 4 bits, and 10 seconds have passed
# the total complexity of this system is expected to be:
# C(10) = 4*10 = 40

# Well, we have calculated how much bits our Universe has. 
# How fast does the complexity of our Universe grow?
```

```python
# In order to get the Growth of complexity, we take the first
# derivative of this equation to get:
# Growth = dC(t)/dt = d(Bits*Time)/dt = Bits

# So system of N bits grows new N connections per every unit of time

# This is how much units of Complexity will be added for a unit of
# time. Here is where we add the assumption of our theory:
# - Every unit of complexity = Unit of 3D space
# - Growth of Complexity = Growth of Volume of Space

# How much volume is generated per unit of time:
# Growth = Universe Bits [VolumeUnits/TimeUnits]

# But we have to be careful, this growth is not yet in meters and
# seconds, it is in fundamental Plank units (Plank Volume and Time)
# - To convert it to SI units (meters and seconds), we multiply
#   by Plank Volume and divide by Plank Time, to get the growth
#   of space in meters^3 / seconds:
Growth = UB*Vp/Tp #[m^3/sec]

# And calculating the answer...
print("The growth rate of space is:", Growth, "[m^3/sec]")
```

```python
# PREDICTING THE HUBBLE EXPANSION RATE

# Lets see what will be the Hubble exp rate in our model
# and compare it with the real (observed) Hubble exp rate

# - But what is the Hubble exp rate (H)?
#   It is how much faster galaxies move away from us as a function of 
#   their distance from us. And here is its observed value:
H = 2.35E-18 #[(m/sec)/m]

# Now we want to predict the Hubble exp rate in out model Universe,
# but how do we do that?
# - Since the space in our model expands with a fixed volume,
#   per every second (this is exactly the Growth we calculated),
#   one can think of how much space gets added for a second between
#   points.
# - Or another way to get to the Hubble constant is by just analyzing
#   the units:
#
#   Hubble exp rate [(m/sec)/m]
#   Growth of space [m^3/sec]
#   
#   to get from [m^3/sec] to [(m/sec)/m] we need to divide
#   by m^3, by volume, by the Universe Volue at the given time

# Long story short, the predicted Hubble exp rate at the current
# moment would be:

pH = Growth / UV
# where UV = Observed Universe Volume

print("Predicted Hubble exp rate:", pH, "[m/sec*m]")
```

```python
# Lets compare this to what the observed Hubble exp rate is:
print("Predicted:", pH)
print("Observed :", H)
```

```python
# NOTE that the observed Hubble rate is with error +2.1 / -2.3
# so our prediction is well within the error margin!

# Our model just exactly predeicted the Hubble exp rate!
```

```python
# PLOTTING THE UNIVERSE RADIUS AS A FUNCITON OF TIME
# - Lets plot the Universe Radius against time and compare the
#   resulting line with the current standard model (Inflation and
#   Big Bang Theory)

# If you ae not interested in the details, you can skip to the
# plot, as it takes a few steps to get there.

# Let's plot time from t=1e-45 to t=1e+35 seconds, in 9 timestamps.
# NOTE: The current Universe Life is about 1e+17 seconds
#       so we are going to plot the past and the future of our
#       Universe.

# - Creating 9 timestamps:
import numpy as np
Timestamps = []
n = -45
while n <= 35:
    timestamp = 10**n
    Timestamps.append(timestamp)
    n+=10
Timestamps = np.array(Timestamps)
```

```python
# - What is the radius as per the standard theoryes, for the given 9 timestamps:
standard_UR  = [1e-55, 
                1e-50,
                1e+10,
                1e+14,
                1e+18,
                1e+22,
                1e+26,
                1e+30,
                1e+34]
standard_UR = np.array(standard_UR)

# NOTE: Here the first two entries are from the pre-Inflationary era,
#       then the Universe went from 1e-50 to 1e+13 Radius (crazy fast!)
#       in a fraction of the second (Inflation), and after that
#       it increases exponetially as per the Big Bang Theory
```

```python
# - What is the predicted Universe Radius as per our theory:
#   We know that the predicted Universe Volume at a given time is the Growth*Time
#   From the Volume we can easily recover the Radius, using
#   the math formula for Radius as a function of Volume:
#   Radius = (3*Volume/4*pi)^(1/3)

# Replacing the Volume with Growth*Time we get:
#   pUR(Time) = (3*(Growth*Time)/4*pi)^(1/3)

#   And now lets calculate the values given out timestamps:
#   where pUR stands for the predicted Universe Radius

predicted_UR = (3*(Growth*Timestamps)/(4*pi))**(1/3)
print("Predicted Radiuses(Timestamps):")
print(predicted_UR)
print("")
```

```python
# - Lets format the data in sci format (10^n),
#   so that we can take a look at the numbers, before we plot them:

sci_format_Timestamps = []
for n in range(len(Timestamps)):
    sci_format_Timestamps.append("{:.2E}".format(Timestamps[n]))
sci_format_pred_UR = []
for n in range(len(predicted_UR)):
    sci_format_pred_UR.append("{:.2E}".format(predicted_UR[n]))
sci_format_st_UR = []
for n in range(len(standard_UR)):
    sci_format_st_UR.append("{:.2E}".format(standard_UR[n]))
    
    
# - Now lets compare the Prediction with the Standard Model:
print("- Given timestamps[sec]:")
print(sci_format_Timestamps)
print("")
print("- Standard Model Radius(t):")
print(sci_format_st_UR)
print("")
print("- Predicted Radius(t):") 
print(sci_format_pred_UR)
```

```python
# - Even before plotting you can see how much smoother solution
#   our model already gives us

# PLOTTING THE DATA:

# - Importing the needed libraries:
import matplotlib
import matplotlib.pyplot as plt
from math import log
%matplotlib inline

# - Convert to logs, since everything is exponential:
log_time         = np.log10(Timestamps.astype(float))
log_predicted_UR = np.log10(predicted_UR.astype(float))
log_standard_UR  = np.log10(standard_UR.astype(float))

# - Set the labels:
plt.xlabel('log of time')
plt.ylabel('log of radius')

# - Create the lines:
plt.plot(log_time, 
         log_predicted_UR, 
         color="green",)
plt.plot(log_time, 
         log_standard_UR, 
         color="red")

# - Plot a dot for the current moment, with the predicted Universe Radius:
# --  current Universe Life:
pX = [UL]        
# -- log of the current Universe Life:
pX = np.log10(pX)
# -- predicted Radius at time = current Universe Life:
pY = (3*(Growth*UL)/(4*pi))**(1/3) 
# -- log of predicted Radius at time = current Universe Life:
pY = np.log10(pY)   
# -- drawing the dot on the plot:
plt.scatter(pX, pY, c="blue", s=100, alpha=0.75)


# NOTE:
# - You need to have some intuition with 
#   the logarithmoc scales in order to
#   to grasp the evolution of the Universe
#   by this graph.
# - If you don't have that, just google "logarithmic scales"
#   or check out this 10 min video as a starter: 
# URL: https://www.khanacademy.org/math/algebra2/exponential-and-logarithmic-functions/logarithmic-scale/v/logarithmic-scale


# Now lets plot those values to compare:
# - Prediction:           Green
# - Standard Model:       Red
# - Universe right now:   the Blue Dot
```
